```python
#Import Pandas Library
import pandas as pd
```

```python
#Store filepath in Variable

film_file = "Resources/film.csv"

#Create a Dataframe by reading CSV using Pandas Library

film_file_df = pd.read_csv(film_file, encoding="ISO-8859-1")

#Display Dataframe's first five lines

film_file_df.head()
    
```

```python
#Display Columns
film_file_df.columns
```

```python
#Rename Columns
organized_film_df = film_file_df.rename(columns={"film_id":"Film_ID", "title": "Title", 'description':'Description', 'release_year':'Release_Year','language_id':'Language_ID',
       'original_language_id':'Original_Language_ID', 'rental_duration':'Rental_Duration', 'rental_rate':'Rental_Rate', 'length':'Length',
       'replacement_cost':'Replacement_Cost', 'rating':'Rating', 'last_update':'Last_Update', 'special_features':'Special_Features',
       'fulltext':'Fulltext'})

organized_film_df.head()
```

```python
 #Display Data Types
organized_film_df.dtypes
```

```python
# Display Unique Values in Columns

# organized_film_df['Language_ID'].unique()
organized_film_df['Original_Language_ID'].unique()
```

```python
#Drop Language Columns
organized_film_df = organized_film_df.drop(columns=['Language_ID', 'Original_Language_ID','Last_Update','Special_Features','Fulltext'])
organized_film_df.head()
```

```python
# Display Data Types
organized_film_df.dtypes
```

```python
# Describe Data
organized_film_df.describe()
```

```python
# Set New Index to Film ID
organized_film_df = organized_film_df.set_index("Title")
organized_film_df.head()
```

```python
#Use loc to show all rows for Title, Rating and Length columns.
organized_film_df = organized_film_df.loc[:, ["Rating","Length"]]
organized_film_df.head()

```

```python
# Use conditional to sort kid friendly films.
Kid_Friendly = organized_film_df["Rating"]=="G"
Kid_Friendly
```

```python
#Use conditional filter to create new DF showing only kid friendly films.
Only_Kid_Friendly_DF = organized_film_df.loc[organized_film_df["Rating"]=="G",:]
Only_Kid_Friendly_DF
```

```python
#Convert new DF to csv, save csv
Only_Kid_Friendly_DF.to_csv(r'Resources\KidFriendlyFilms.csv')
```
